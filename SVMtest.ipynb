```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_circles
```

```python
# Creating Data with some noises
X, y = make_circles(n_samples=300, factor=0.5, noise=0.12, random_state=42)
```

```python
# Plt show the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=50)
plt.title("Data=")
plt.show()
```

```python
# Training the Data with RBF model
model = SVC(kernel='rbf', C=1.0)
model.fit(X, y)
```

```python
# Showing the Hyperplane and Margins
xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 200),
                     np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 200))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=50, label="Data")
plt.contour(xx, yy, Z, levels=[0], colors='k', linewidths=2, label="Hyperplane")
plt.contour(xx, yy, Z, levels=[-1, 1], colors='gray', linestyles='--', linewidths=1.5, label="Soft margin")
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], edgecolor='k', facecolor='none', s=100, label="Support vectors")
plt.title("Data after SVM")
plt.legend() #idk why i did plt.legend but still having warnings (these are harmless warnings)
plt.xlim(-1.3, 1.3)
plt.ylim(-1.3, 1.3)
plt.show()
```

```python
# New point prediction
new_point = np.array([[0.2, 0.3]])  # New point that needed to be predict
predicted_class = model.predict(new_point)
print(f"Point {new_point[0]} predicted_class: {predicted_class[0]}")
```
